````python
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time

# For Windows
plt.rc('font', family='gulim') 
plt.rc('axes', unicode_minus=False)

## 검색창에 검색어 입력
def searching_news(coin : str):
    global titledata, keyword, press
    keyword = coin
    if press == None:
        dfname = keyword
    else:
        dfname = keyword+'_'+press
    search_bar = driver.find_element('xpath', '//*[@id="q"]')
    search_bar.click()
    search_bar.clear()
    search_bar.send_keys('\"'+keyword+'\"')
    search_bar.send_keys(Keys.RETURN)
    if not dfname in titledata:
        titledata[dfname] = pd.DataFrame(columns=['date', 'title', 'press'])
    driver.implicitly_wait(3)

## 최신순 정렬
def set_newest():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[1]/div[1]/a[2]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (조선일보)
def switch_to_chosun():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[9]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='조선일보']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (중앙일보)
def switch_to_joongang():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[9]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='중앙일보']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (동아일보)
def switch_to_donga():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[3]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='동아일보']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (한겨레)
def switch_to_hani():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[14]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='한겨레']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (토큰포스트)
def switch_to_tokenpost():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[12]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='토큰포스트']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)

## 언론사 교체 (코인데스크코리아)
def switch_to_coindesk():
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/a/span[2]')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[2]/ul/li[11]/a')
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[3]/ul')
    pointer = pointer.find_element('xpath', "//*[text()='코인데스크코리아']")
    pointer.click()
    time.sleep(0.1)
    pointer = driver.find_element('xpath', '//*[@id="newsColl"]/div[1]/div[2]/div/div[2]/div/div[4]/button[1]')
    pointer.click()
    driver.implicitly_wait(3)




## 최대 100개의 기사 제목 수집(temp=30)
def collecting_headlines():
    global titledata, keyword, press
    if press == None:
        dfname = keyword
    else:
        dfname = keyword+'_'+press
    for _ in range(3):
        elems = driver.find_elements('xpath', "//*[@class='list_news']/li")
        for elem in elems:
            title = elem.find_element('class name', 'tit_main')
            postingdate = elem.find_element('xpath', "//span[@class='f_nb']")
            npname = elem.find_element('xpath', "//a[@class='f_nb']")
            titledata[dfname].loc[len(titledata[dfname])] = [postingdate.text, title.text, npname.text]
        # 2페이지가 없는 경우 리턴
        try:
            pointer = driver.find_element('class name', 'btn_next')
        except:
            return
        # 마지막 페이지일 경우 리턴
        if pointer.get_attribute("data-paging-active") == 'false':
            return
        pointer.click()
        time.sleep(1.1)
        driver.implicitly_wait(3)


## 최대 100개의 중앙일보 기사 수집(temp=30)
def collecting_joongang():
    # global titledata, keyword, press
    # if press == None:
    #     dfname = keyword
    # else:
    #     dfname = keyword+'_'+press
    global articledata
    articledata[0] = pd.DataFrame(columns=['press', 'title', 'headline', 'date', 'reporter','article', 'page'])
    for _ in range(10):
        # 뉴스박스 선택
        elems = driver.find_elements('xpath', "//*[@class='list_news']/li")
        for elem in elems:
            elem.find_element('class name', 'tit_main').click()
            last_tab = driver.window_handles[-1]
            driver.switch_to.window(window_name=last_tab)
            driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') 
            driver.implicitly_wait(5)
            
            title = ''
            try:
                title += driver.find_element('xpath', "//*[@id='container']/section/article/header/div[1]/a").text
                title += driver.find_element('xpath', "//*[@id='container']/section/article/header/div[1]/span").text
            except:
                pass
            
            headline = driver.find_element('xpath', "//*[@id='container']/section/article/header/h1").text
            
            date = driver.find_element('xpath', "//*[@id='container']/section/article/header/div[2]/div[1]/p").text
            
            try:
                reporter = driver.find_element('xpath', "//*[@id='container']/section/article/header/div[3]/a").text
            except:
                reporter = '외부사설'
            
            ps = driver.find_elements('xpath', "//*[@id='article_body']/p")
            article = ''
            for p in ps:
                article += p.text
            
            page = driver.current_url
            
            articledata[0].loc[len(articledata[0])] = ['joongang', title, headline, date, reporter, article, page]
            driver.close()
            last_tab = driver.window_handles[-1]
            driver.switch_to.window(window_name=last_tab)
        # 2페이지가 없는 경우 리턴
        try:
            pointer = driver.find_element('class name', 'btn_next')
        except:
            return
        # 마지막 페이지일 경우 리턴
        if pointer.get_attribute("data-paging-active") == 'false':
            return
        pointer.click()
        time.sleep(1.1)
        driver.implicitly_wait(3)



def switch_to(cpname):
    if cpname == '조선':
        switch_to_chosun()
    elif cpname == '중앙':
        switch_to_joongang()
    elif cpname == '동아':
        switch_to_donga()
    elif cpname == '한겨례':
        switch_to_hani()
    elif cpname == '토큰포스트':
        switch_to_tokenpost()
    elif cpname == '코인데스크':
        switch_to_coindesk()
    else:
        return(False)

## 검색어로 찾은 최대 100개의 뉴스 제목을 수집해 데이터프레임에 저장합니다.
def create_df(coin : str, cpname = None):
    global titledata, press
    press = cpname
    if press == None:
        dfname = coin
    else:
        dfname = coin+'_'+cpname
    searching_news(coin)
    set_newest()
    if cpname != None:
        if switch_to(cpname)==False:
            return
    collecting_headlines()
    print(titledata[dfname])
````
